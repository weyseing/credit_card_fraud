## Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Dataset

```python
df = pd.read_csv(os.getcwd() + '/dataset/clean_creditcard.csv')
df.tail(3)
```

## Visual

```python
fnfTransactions = df["Class_Category"].value_counts()
fnfTransactions
```

```python
plt.xlabel("Class")
plt.ylabel("Number of Observations")
fnfTransactions.plot(kind = 'bar',title = 'Frequency by observation number')
plt.show()
```

```python
plt.xlabel("Class")
plt.ylabel("Number of Observations")
fnfTransactions.plot(kind = 'pie',title = 'Frequency by observation number')
plt.show()
```

```python
df['Stan_Amount'].hist()
plt.xlabel("Standarized Amount")
plt.ylabel("Number of Observations")
plt.show()
```

```python
df['Time'].hist()
plt.xlabel("Time")
plt.ylabel("Number of Observations")
plt.show()
```

```python
ax = df.boxplot(column=["Stan_Amount"], by="Class_Category", grid=False)
ax.set_xlabel("Class_Category")
ax.set_ylabel("Standarized Amount")
plt.show()
```

```python
plt.scatter(df.loc[df['Class_Category'] == 0]['V1'], df.loc[df['Class_Category'] == 0]['V2'],alpha=0.5, linewidth=0.15)
plt.scatter(df.loc[df['Class_Category'] == 1]['V1'], df.loc[df['Class_Category'] == 1]['V2'],alpha=0.5, linewidth=0.15,c='r')
plt.show()
```
